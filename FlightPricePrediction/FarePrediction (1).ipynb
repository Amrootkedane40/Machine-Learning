```python
#Removing correlated features
Threshold=0.9
# find and remove correlated features
def correlation(dataset, threshold):
    col_corr = set()  # Set of all the names of correlated columns
    corr_matrix = dataset.corr()
    for i in range(len(corr_matrix.columns)):
        for j in range(i):
            if abs(corr_matrix.iloc[i, j]) > threshold: # we are interested in absolute coeff value
                colname = corr_matrix.columns[i]  # getting the name of column
                col_corr.add(colname)
    return col_corr

correlation(X,Threshold)
```

### Using Tobit Regressor

```python
import statsmodels.api as sm
from scipy.stats import norm
from sklearn.metrics import r2_score

X_with_constant = sm.add_constant(X)


censored_y = np.where(y == 0, np.nan, y)
is_censored = np.isnan(censored_y)
model = sm.OLS(censored_y, X_with_constant)


censored_results = model.fit()

mu = censored_results.predict(X_with_constant)
sigma = censored_results.scale

z = (y - mu) / sigma
truncated_z = np.where(is_censored, norm.ppf(norm.cdf(z)), z)
truncated_y = mu + truncated_z * sigma


tobit_model = sm.OLS(truncated_y, X_with_constant)
tobit_results = tobit_model.fit()
r2 = r2_score(truncated_y, tobit_results.predict(X_with_constant))
print("R2 Score:", r2)

#print(tobit_results.summary())
```

```python
import statsmodels.api as sm
from scipy.stats import norm
from sklearn.metrics import mean_squared_error, mean_absolute_error

X_with_constant = sm.add_constant(X)

censored_y = np.where(y == 0, np.nan, y)
is_censored = np.isnan(censored_y)

model = sm.OLS(censored_y, X_with_constant)
censored_results = model.fit()

mu = censored_results.predict(X_with_constant)
sigma = censored_results.scale

z = (y - mu) / sigma
truncated_z = np.where(is_censored, norm.ppf(norm.cdf(z)), z)
truncated_y = mu + truncated_z * sigma

tobit_model = sm.OLS(truncated_y, X_with_constant)
tobit_results = tobit_model.fit()

predicted_values = tobit_results.predict(X_with_constant)

rmse = np.sqrt(mean_squared_error(truncated_y, predicted_values))
mse = mean_squared_error(truncated_y, predicted_values)
mae = mean_absolute_error(truncated_y, predicted_values)

print("RMSE:", rmse)
print("MSE:", mse)
print("MAE:", mae)
```
